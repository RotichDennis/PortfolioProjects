```python
# Importing of packages

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Reading the file
df = pd.read_csv(r'C:\Users\rotic\Downloads\movies.csv')
```

```python
df
```

```python
#Checking for any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Checking the different datatypes with using their columns
print(df.dtypes)
```

```python
#Checking for outliers
df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
#order our data a little to see
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#scatterplot for the gross function
sns.regplot(x="gross", y="budget", data=df)
```

```python
#scatterplot for scores against their gross
sns.regplot(x="score", y="gross", data=df)
```

```python
#Correlation Matrix on all numeric columns..others like peasorn and kendall can be added
df.corr()
```

```python
#correlation tables of the numeric columns using correlation matrix
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Looking at the top 15 companies by gross revenue
CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
#Separating month with the released time
df['Month'] = df['released'].astype(str).str[:4]
df
```

```python
#company, year and gross sum
CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
#Scatterplot for budget and gross
plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```
